```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

# Data
weather = ["sunny", "sunny", "overcast", "rain", "rain", "sunny"]
play = ["no", "no", "yes", "yes", "yes", "yes"]

# Encode categorical data
le_weather = LabelEncoder()
le_play = LabelEncoder()
weather_encoded = le_weather.fit_transform(weather).reshape(-1, 1)
play_encoded = le_play.fit_transform(play)

# Decision Tree
model1 = DecisionTreeClassifier()
model1.fit(weather_encoded, play_encoded)
pred1 = model1.predict(le_weather.transform(["sunny"]).reshape(-1, 1))
print("Decision Tree:", le_play.inverse_transform(pred1))

# KNN
model2 = KNeighborsClassifier(n_neighbors=3)
model2.fit(weather_encoded, play_encoded)
pred2 = model2.predict(le_weather.transform(["sunny"]).reshape(-1, 1))
print("KNN:", le_play.inverse_transform(pred2))

# Naive Bayes
model3 = GaussianNB()
model3.fit(weather_encoded, play_encoded)
pred3 = model3.predict(le_weather.transform(["sunny"]).reshape(-1, 1))
print("Naive Bayes:", le_play.inverse_transform(pred3))
```
